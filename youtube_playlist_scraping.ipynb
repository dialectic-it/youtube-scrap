# Get video_info from video_url:

- video_duration
- video_date
- video_title

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

df = pd.read_excel('youtube_videos.xlsx')
df.head()
```

```python
import yt_dlp
import time

# https://github.com/yt-dlp/yt-dlp/issues/3873
# https://github.com/yt-dlp/yt-dlp/wiki/FAQ#how-do-i-pass-cookies-to-yt-dlp
ydl_opts = {'quiet': True,
            'verbose': False,
            #'cookies-from-browser': 'chrome', # не сработало :с
            #'cookies': './cookies.txt', # не сработало :с
           }


data = []
for i in range(0,len(df)):
    video_url = df['video_url'][i]
    print(i, video_url)
    time.sleep(3)
    
    try:
        with yt_dlp.YoutubeDL(ydl_opts) as yt:
            meta = yt.extract_info(video_url, download=False)
        data.append({'video_url':video_url,
                     'duration':meta['duration'],
                     'upload_date':meta['upload_date'],
                     'title':meta['title']})   
    except: # ошибки для видео, которым выдали ограничения
        data.append({'video_url':video_url,
                     'duration':'error',
                     'upload_date':'error',
                     'title':'error'})         
    print(data[-1], '\n')
    
df1 = pd.DataFrame(data)
#df1.to_excel('youtube_videos_meta.xlsx', index=False)
```

```python
df['video_duration'] = df1['duration']
df['video_date'] = df1['upload_date']
df['video_title'] = df1['title']

df = df[['video_id','video_duration','video_date','video_url','video_title',
         'channel_url','playlist_url','playlist_name']]
#df.to_excel('yt_videos_info.xlsx', index=False)
df.head(3).T
```
